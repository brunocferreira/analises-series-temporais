### 1. Defina o que é uma série temporal e explique como ela se difere de outros tipos de dados.


Uma **série temporal** é uma sequência de pontos de dados ordenados no tempo. Normalmente, a série temporal possui uma sequência de dados equidistantes no tempo. Os dados de série temporal são uma coleção de quantidades reunidas em intervalos regulares de tempo e ordenadas cronologicamente.

As séries temporais são diferentes de outros tipos de dados, pois têm uma dimensão temporal. Isso significa que os dados são coletados ou gerados em intervalos regulares de tempo, e esses intervalos são importantes para a análise dos dados.

A análise de séries temporais é realizada com o intuito de explorar o comportamento passado e também de prever o comportamento futuro em um determinado problema. As séries temporais são amplamente utilizadas em campos como meteorologia, economia, finanças, epidemiologia e física.

As séries temporais são usadas em uma variedade de aplicações, incluindo:

1. **Previsão**: as séries temporais podem ser usadas para prever valores futuros. Por exemplo, uma série temporal de vendas pode ser usada para prever as vendas futuras.
2. **Análise de tendências**: as séries temporais podem ser usadas para identificar tendências ao longo do tempo. Por exemplo, uma série temporal de temperatura pode ser usada para identificar tendências climáticas.
3. **Análise de sazonalidade**: as séries temporais podem ser usadas para identificar padrões sazonais. Por exemplo, uma série temporal de vendas pode ser usada para identificar padrões de vendas sazonais.

As principais diferenças entre séries temporais e outros tipos de dados:

1. **Dimensão temporal**: séries temporais têm uma dimensão temporal, o que significa que os dados são coletados ou gerados em intervalos regulares de tempo. Outros tipos de dados não têm essa dimensão temporal.
2. **Padrão**: séries temporais geralmente apresentam padrões, como tendências, sazonalidade e ciclos. Outros tipos de dados podem apresentar padrões, mas não são tão comuns quanto em séries temporais.
3. **Análise**: séries temporais são geralmente analisadas usando métodos específicos para séries temporais. Outros tipos de dados podem ser analisados usando uma variedade de métodos.

Algumas diferenças entre as séries temporais e outros tipos de dados:

1. **Dados transversais**: dados que são coletados em um único ponto no tempo. Por exemplo, uma pesquisa de opinião que é realizada em um único dia é um conjunto de dados transversais. Os dados transversais são geralmente usados para descrever um fenômeno em um determinado momento.
2. **Dados do painel**: dados que são coletados em vários pontos no tempo para o mesmo grupo de indivíduos ou objetos. Por exemplo, um estudo que acompanha o desempenho de um grupo de alunos ao longo do tempo é um conjunto de dados do painel. Os dados do painel são geralmente usados para identificar mudanças ao longo do tempo.
3. **Dados OLTP (processamento de transações online)**: dados que são coletados e processados em tempo real. Por exemplo, os dados de transações de um banco são dados OLTP. Os dados OLTP são geralmente usados para operações de negócios cotidianas.
4. **Dados categóricos**: dados categóricos são dados que são classificados em categorias. Por exemplo, a cor de um carro pode ser classificada nas categorias vermelho, azul, verde, etc. Os dados categóricos não têm uma dimensão temporal.
5. **Dados numéricos**: dados numéricos são dados que são representados por números. Por exemplo, a altura de uma pessoa pode ser representada pelo número 1,81 metros. Os dados numéricos podem ter uma dimensão temporal, mas não é necessário. Por exemplo, a altura de uma pessoa pode ser medida uma única vez, sem intervalos de tempo regulares.
6. **Dados de texto**: dados de texto são dados que são representados por texto. Por exemplo, um artigo de jornal pode ser representado por texto. Os dados de texto não têm uma dimensão temporal.

### 3. Suponha que você tenha dados de vendas diárias de um produto ao longo de um ano. Descreva três etapas que você seguiria para realizar uma análise inicial desses dados de séries temporais.


1. **Etapa 1: _Preparação dos dados_**

A primeira etapa da análise de séries temporais é a preparação dos dados. Os dados de séries temporais podem conter erros, valores ausentes ou outras anomalias. A preparação dos dados ajuda a garantir que os dados sejam precisos e confiáveis.

Tarefas comuns de preparação de dados para séries temporais:

* **Remoção de erros**: os dados de séries temporais podem conter erros, como valores inválidos ou incorretos. Os erros podem ser identificados usando técnicas de análise de dados, como análise de outliers ou análise de resíduos.

* **Remoção de valores ausentes**: os dados de séries temporais podem conter valores ausentes, o que pode afetar a precisão das análises. Os valores ausentes podem ser removidos, substituídos por valores estimados ou imputados usando técnicas de aprendizado de máquina.

* **Transformação dos dados**: os dados de séries temporais podem ser transformados para melhorar a precisão das análises. As transformações comuns incluem normalização, transformação logarítmica e transformação de Box-Cox.

Aplicação às vendas diárias de um produto:

No caso das vendas diárias de um produto, as seguintes tarefas de preparação de dados podem ser realizadas:

* _Verificação de erros_: os dados podem ser verificados para garantir que não contenham valores inválidos ou incorretos. Por exemplo, as vendas não podem ser negativas ou zero.

* _Remoção de valores ausentes_: os valores ausentes podem ser removidos se forem poucos. Se os valores ausentes forem muitos, eles podem ser substituídos por valores estimados ou imputados usando técnicas de aprendizado de máquina.

* _Transformação dos dados_: as vendas podem ser transformadas para melhorar a precisão das análises. Por exemplo, a transformação logarítmica pode ser usada para reduzir a variabilidade dos dados.

2. **Etapa 2: _Visualização dos dados_**

Após a preparação dos dados, a próxima etapa é a visualização dos dados para obter uma compreensão geral do comportamento das vendas. Isso pode ser feito criando um gráfico de linha dos dados. O gráfico de linha deve mostrar as vendas ao longo do tempo, com cada ponto de dados representando uma data diferente.

Aplicação às vendas diárias de um produto:

* Um gráfico de linha das vendas diárias de um produto pode mostrar padrões ou tendências nas vendas. Por exemplo, o gráfico pode mostrar que as vendas são mais altas durante certos períodos do ano, como o Natal ou o Dia dos Namorados.

* A verificação da estacionaridade da série temporal deve ser realizada antes de prosseguir com as etapas de análise. A estacionaridade é uma propriedade importante das séries temporais que garante que os modelos de previsão sejam robustos e precisos.

> _Verificação da estacionariedade dos dados_:
> * **Visualização**: um gráfico de linha das vendas diárias de um produto pode mostrar se a média e a variância das vendas são constantes ao longo do tempo. Se a média e a variância forem constantes, a série temporal é provavelmente estacionária.
> * **Testes estatísticos**: os testes de Dickey-Fuller e KPSS podem ser usados para verificar a presença de uma raiz unitária ou autocorrelação de longo prazo nas vendas diárias de um produto. Se os testes forem significativos, a série temporal não é estacionária.
> * **Transformações**: se as vendas diárias de um produto não forem estacionárias, elas podem ser transformadas usando uma transformação da diferença de ordem N, transformção logarítmica ou Box-Cox. A transformação logarítmica é frequentemente usada para reduzir a variância das vendas diárias de um produto.

A terceira etapa é realizar uma análise estatística dos dados para obter insights mais profundos. Isso pode incluir calcular medidas estatísticas, como média, mediana, desvio padrão e coeficiente de variação.

* #### Tendência
##### **Conjunto de dados da temperatura da superfície da terra de 1832 à 2013**
Para o estudo iremos delimitar apenas a superfície do Brasil

```python
# Importa e apresentar os dados
temp_data = pd.read_csv("./dataset_tendencia/GlobalLandTemperaturesByCountry.csv")
temp_data.head()
```

```python
# Capturar apenas os valores do Brasil
dados = temp_data.loc[temp_data['Country'] == 'Brazil', ['dt', 'AverageTemperature', 'Country']]
```

```python
dados.head()
```

```python
# Apresentar esses valores em um gráfico
dados.plot()
plt.legend(loc='best')
plt.show()
```

```python
# Separar apenas os valores médios da temperatura
dados = dados["AverageTemperature"]
```

```python
# Verifica os dados
print(dados.head())
print(dados.info())
```

```python
# Cria um novo índice de data
date_index = pd.date_range(start='01/01/1832', end='01/09/2013', freq='M')

# Se a série é mais longa que o índice de data, trunca a série
if len(dados) > len(date_index):
    dados = dados.iloc[:len(date_index)]

# Se o índice de data é mais longo que a série, preenche os valores faltantes com NaN
elif len(date_index) > len(dados):
    dados = dados.reindex(date_index)

# Atribui o novo índice à série
dados.index = date_index

# Verifica as alterações nos índices dos dados
print(dados.head())
print(dados.info())
```

```python
# Apresentar os dados originais da série e a média móvel
media_movel = dados.rolling(window=12).mean()
plt.plot(dados, label='Série Real')
plt.plot(media_movel,color='red', label='Média Móveis')
plt.legend(loc='best')
plt.show()
```

```python
# Remove os valores NaN (apenas o último registro de 01/09/2013 está ausente)
media_movel = media_movel.dropna()

# Cria um array com as datas da série
X = np.array(media_movel.index).reshape(-1, 1)

# Converte as datas para números (necessário para o modelo de regressão linear)
X = X.astype(np.int64) // 10**9

# Cria um modelo de regressão linear
model = LinearRegression()

# Ajusta o modelo aos dados
model.fit(X, media_movel.values)

# Usa o modelo para prever a tendência
tendencia = model.predict(X)

# Plota a série original, a média móvel e a tendência
plt.plot(dados.index, dados, label='Série Real')
plt.plot(media_movel.index, media_movel, color='red', label='Média Móveis')
plt.plot(media_movel.index, tendencia, color='purple', label='Tendência')
plt.legend(loc='best')
plt.show()
```

```python
# Captura o primeiro valor da tendência
primeiro_valor = tendencia[0]

# Captura o último valor da tendência
ultimo_valor = tendencia[-1]

# Imprime os valores
print(f"Valor médio da temperatura entre 1832 e 1833: {primeiro_valor}")
print(f"Valor médio da temperatura entre 2012 e 2013: {ultimo_valor}")
```

##### **Conclusão**: A partir dos dados coletados pelo Berkeley e da análise que fizemos, podemos afirmar que sim, existe uma tentência de aumento da temperatura do Brasil ao longo do tempo a partir da amostragem dos dados.

* #### Sazonalidade
##### **Conjunto de dados das vendas de sorvete de 01/01/2019 à 31/12/2019**
Na base do kaggle não havia mais informações sobre os dados na descrição.

```python
# Importa e apresentar os dados
icecream_data = pd.read_csv("./dataset_sazonalidade/icecream_sales.csv")
icecream_data.head()
```

```python
# Converte 'date' para o formato de data
icecream_data['date'] = pd.to_datetime(icecream_data['date'])

# Define 'date' como índice e seleciona apenas a coluna 'sales'
dados = icecream_data.set_index('date')['sales']

# Imprimir os dados e o tipo de dados da série
print(dados.head())
print(dados.info())
```

```python
# Define a frequência da série temporal
dados = dados.asfreq('D')

# Preenche os valores NaN restantes com o próximo valor na série
dados = dados.fillna(method='bfill')

# Preenche os valores NaN com o valor anterior na série
dados = dados.fillna(method='ffill')

# Realiza a decomposição sazonal
decomposicao = sm.tsa.seasonal_decompose(dados, model='additive')

# Dividir a decomposição
tendencia = decomposicao.trend
sazonalidade = decomposicao.seasonal
residuo = decomposicao.resid

# Plotar os componentes da decomposição
plt.subplot(411)
plt.plot(dados, label='Original', color='red')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(tendencia, label='Tendência', color='green')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(sazonalidade,label='Sazonalidade', color='blue')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residuo, label='Resíduos', color='orange')
plt.legend(loc='best')
plt.tight_layout()
plt.show()


```

```python
# Imprimindo apenas a sazonalidade
plt.plot(sazonalidade,label='Sazonalidade', color='blue')
plt.legend(loc='best')
plt.show()
```

```python
# Imprime apenas a sazonalidade
sazonalidade.head()
```

```python
# Converte a série em um DataFrame
sazonalidade = sazonalidade.to_frame()

# Cria uma nova coluna 'é_final_de_semana' que é True se a data é um final de semana e False caso contrário
sazonalidade['é_final_de_semana'] = sazonalidade.index.weekday >= 5

# Agrupa os dados por 'é_final_de_semana' e capitura o valor máximo da sazonalidade
media_vendas_por_dia = sazonalidade.groupby('é_final_de_semana')['seasonal'].max()

print(media_vendas_por_dia)
```

##### **Conclusão**: A partir dos dados disponíveis no kaggle e da análise que fizemos, podemos afirmar que sim, existe uma sazonalidade nas vendas de sorvete ao longo dos dias do ano com uma concentração maior das vendas nos finais de semana a partir da amostragem dos dados.

```python
# Importa e apresentar os dados
poluicao_data = pd.read_csv("./dataset_estacionariedade/a6_poluicao.csv")
poluicao_data.head()
```

```python
# Converte 'date' para o formato de data
poluicao_data['date'] = pd.to_datetime(poluicao_data['DATA'])

# Define 'date' como índice e seleciona apenas a coluna 'dióxido de nitrogênio'
dados = poluicao_data.set_index('date')['no2']

# Imprimir os dados e o tipo de dados da série
print(dados.head())
print(dados.info())
```

```python
# Define a frequência da série temporal
dados = dados.asfreq('D')

# Preenche os valores NaN restantes com o próximo valor na série
dados = dados.fillna(method='bfill')

# Preenche os valores NaN com o valor anterior na série
dados = dados.fillna(method='ffill')

# Realiza a decomposição sazonal
decomposicao = sm.tsa.seasonal_decompose(dados, model='additive')

# Dividir a decomposição
tendencia = decomposicao.trend
sazonalidade = decomposicao.seasonal
residuo = decomposicao.resid

# Plotar os componentes da decomposição
plt.subplot(411)
plt.plot(dados, label='Original', color='red')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(tendencia, label='Tendência', color='green')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(sazonalidade,label='Sazonalidade', color='blue')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residuo, label='Resíduos', color='orange')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Realiza o teste Dickey-Fuller
resultado = sm.tsa.stattools.adfuller(dados)

print(f'ADF Statistic: {resultado[0]}')
print(f'p-value: {resultado[1]}')
```

##### **Conclusão**: A partir dos dados expostos do livro de Análises de Séries Temporais de Pedro A. Moretin, disponíveis no kaggle, juntamente com a análise que fizemos, podemos afirmar que sim, esta é uma série estacionária ao descartamos a hipótese nula devido o ADF Statistic ter sido negativo, um p-valor superiror a 0,05, confirmando a nossa hipótese além da linha de tendência se manter, praticamente, constante e horizontal ao longo tempo a partir da amostragem dos dados.

```python
# Preenche os valores NaN restantes com o próximo valor na série
dados = dados.fillna(method='bfill')

# Preenche os valores NaN com o valor anterior na série
dados = dados.fillna(method='ffill')

# Realiza a decomposição sazonal
decomposicao = sm.tsa.seasonal_decompose(dados, model='additive')

# Dividir a decomposição
tendencia = decomposicao.trend
sazonalidade = decomposicao.seasonal
residuo = decomposicao.resid

# Plotar os componentes da decomposição
plt.subplot(411)
plt.plot(dados, label='Original', color='red')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(tendencia, label='Tendência', color='green')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(sazonalidade,label='Sazonalidade', color='blue')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residuo, label='Resíduos', color='orange')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Realiza o teste Dickey-Fuller
resultado = sm.tsa.stattools.adfuller(dados)

print(f'ADF Statistic: {resultado[0]}')
print(f'p-value: {resultado[1]}')
```

```python
# Calcular a média
media = dados.mean()

# Calcular a variância
variancia = dados.var()

# Imprimir a média e a variância
print('Média:', media)
print('Variância:', variancia)

# Converte o índice de data em uma série
date_series = pd.Series(date_index.map(lambda x : x.value))

# Calcula a correlação
correlacao = dados.corr(date_series)

# Imprimir a matriz de correlação
print(correlacao)

# Cria um gráfico de dispersão
plt.scatter(dados, date_series)

# Adiciona um título e rótulos aos eixos
plt.title('Gráfico de Dispersão entre Dados e Referência')
plt.xlabel('Dados')
plt.ylabel('Referencia')

# Mostra o gráfico
plt.show()
```
